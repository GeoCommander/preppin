```python
import pandas as pd

url='https://drive.google.com/u/0/uc?id=1oln2ri6nu1wDQfT3gQMLLNlmQ2h6B9d9&export=download'

data=pd.read_csv(url)
```

```python
data.head()
```

```python
data[['Bank','Transaction Code']]=data['Transaction Code'].str.split('-',n=1,expand=True)
```

```python
data.head()
```

```python
data=data[['Transaction Code','Bank','Value','Customer Code','Online or In-Person','Transaction Date']]
```

```python
data.head()
```

```python
mapping={1:'Online',2:'In-Person'}
data['Online or In-Person']=data['Online or In-Person'].replace(mapping)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['Transaction Date']=pd.to_datetime(data['Transaction Date'])
```

```python
data.dtypes
```

```python
data['Transaction Date']=data['Transaction Date'].dt.strftime('%A')
```

```python
data.head()
```

```python
by_bank=data.groupby(['Bank'])['Value'].sum()
by_3_properties=data.groupby(['Bank','Transaction Date','Online or In-Person'])['Value'].sum()
by_2_properties=data.groupby(['Bank','Customer Code'])['Value'].sum()
```

```python
data.to_csv('challenge_1_table.csv',index=False)
by_bank.to_csv('by_bank.csv')
by_3_properties.to_csv('by_3_properties.csv')
by_2_properties.to_csv('by_2_poperties.csv')
```
